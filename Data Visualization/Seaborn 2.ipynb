```python
import seaborn as sns
sns.set() 

# set digunakan agar setting plot beralih ke seaborn, 
# karena secara default seaborn juga dibuat berdasarkan matplotlib
```

```python
sns.get_dataset_names()
```

```python
df_titanic = sns.load_dataset("titanic")
```

```python
df_titanic.head()
```

```python
df_titanic.shape
```

### 1. Distributon plot

```python
sns.distplot(df_titanic.age);
```

```python
sns.distplot(df_titanic.age,
             kde = False); # grafik
```

```python
sns.distplot(df_titanic.age,
             kde = True,
             bins = 40); # jumlah batang
```

### 2. Jointplot

```python
sns.jointplot(x = "age",
             y = "fare",
             data = df_titanic);
```

```python
import scipy.stats as stat
```

```python
sns.jointplot("age", "fare",
             stat_func = stat.pearsonr,
             data = df_titanic);
```

```python
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
sns.jointplot("age", "fare",
             stat_func = stat.pearsonr,
             data = df_titanic,
             kind = "reg");
```

```python
# cek outlier
# akan di drop
# agar tampilan lebih baik
```

```python
df_titanic.fare[df_titanic.fare > 300]
```

```python
titanic_new = df_titanic.drop([258, 679, 737])
```

```python
titanic_new.shape
```

```python
sns.jointplot("age", "fare",
             stat_func = stat.pearsonr,
             data = titanic_new,
             kind = "reg");
```

```python
from scipy.stats import zscore
```

```python
import numpy as np
```

```python
nilai_z = zscore(df_titanic.fare)
nilai_z
```

```python
nilai_z_abs = np.abs(nilai_z)
nilai_z_abs
```

```python
filter_data = nilai_z_abs < 3
```

```python
titanic_new1 = df_titanic[filter_data]
```

```python
titanic_new1.head()
```

```python
titanic_new1.shape
```

```python
sns.jointplot("age", "fare",
             stat_func = stat.pearsonr,
             data = titanic_new1,
             kind = "reg");
```

```python
sns.jointplot("age", "fare",
             stat_func = stat.pearsonr,
             data = titanic_new1,
             kind = "hex");
```

```python
sns.jointplot("age", "fare",
             stat_func = stat.pearsonr,
             data = titanic_new1,
             kind = "kde");
```

```python
sns.heatmap(pivot_flights, cmap = "Reds", 
            linecolor = "white",
           linewidths = 0.5)
```
