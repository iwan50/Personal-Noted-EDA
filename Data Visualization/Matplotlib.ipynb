##### Tambahan Fitur

```python
plt.figure(figsize= (12,9)) # Ukuran plot
plt.plot(x, y, x, z)        # Membuat plot
plt.title("Latihan")        # Judul plot
plt.xlabel("Sumbu X")       # Sumbu X
plt.ylabel("Sumbu Y")       # Sumbu Y
plt.grid(True)              # Menampilkan grid
plt.legend(["Garis XY", "Garis XZ"], loc= 0)
# Menampilkan legend, loc= lokasi legend, 0 adalah posisi terbaik
plt.savefig("plot1.png")    # Save gambar
plt.show()                  # Menghilangkan tulisan "module"
```

```python
x = np.arange(1, 11)
y = x ** 2
z = x ** 3
```

```python
plt.figure(figsize= (12,9)) # Ukuran plot
plt.plot(x, x,"red",
         x, y,"green",
         x, z,"orange") # Membuat plot

plt.title("Latihan")        # Judul plot
plt.xlabel("Sumbu X")       # Sumbu X
plt.ylabel("Sumbu Y")       # Sumbu Y
plt.grid(True)              # Menampilkan grid
plt.legend(["Garis XX", "Garis XY", "Garis XZ"], loc= 0)
# Menampilkan legend, loc= lokasi legend, 0 adalah posisi terbaik

# plt.savefig("plot1.png")    # Save gambar
plt.show()                  # Menghilangkan tulisan "module"
```

##### CUSTOM

```python
plt.figure(figsize= (12,9)) # Ukuran plot
plt.plot(x, y,
         color = "orange",
         marker = ">",
         markersize = 12,
         markerfacecolor = "green",
         markeredgecolor = "yellow",
         linestyle = "--",
         linewidth = 2,
         label = "Pojok Kanan"
        )                   # Membuat plot
plt.title("Latihan")        # Judul plot
plt.xlabel("Sumbu X")       # Sumbu X
plt.ylabel("Sumbu Y")       # Sumbu Y
plt.grid(True)              # Menampilkan grid
plt.legend()# (["Garis XX"], loc= 0)
# Menampilkan legend, loc= lokasi legend, 0 adalah posisi terbaik

# y = [1,2,3,4,5]
# x = [11,12,13,14,15]

for i in y:
    ind = list(y).index(i)
    plt.text(x[ind]-0.2, i+2, i)

plt.annotate("Nilai Karl Max",
            xy = (9,81),
            xytext = (6,90),
             arrowprops = {
                 "facecolor" : "red",
                 "shrink" : 0.1,
                 "width" : 7,
                 "headwidth" : 20,
                 "headlength" : 20
             }
            )

# plt.text(posisi sumbu X, posisi sumbu Y, text)
plt.text(8, 20, "Ini adalah teks")





# plt.savefig("plot1.png")    # Save gambar
plt.show()                  # Menghilangkan tulisan "module"
```

## 2. Bar plot

```python
df1 = pd.DataFrame({
    "x": [1,2,3,4,5],
    "y": [5,4,6,3,5],
    "z": [8,7,9,7,6]
})
```

```python
df1
```

##### Line plot

```python
plt.plot(df1["x"], df1["y"])
```

##### Bar plot

```python
plt.bar(df1["x"], df1["y"])
```

##### Bentuk Horizontal

```python
plt.barh(df1["x"], df1["y"])
```

##### Tambahan Fitur

```python
plt.figure(figsize= (13,9))
plt.bar(df1["x"], df1["y"],
       color = ["red", "green", "orange", "yellow", "black" ],
        linewidth = 3,
        edgecolor = "black"
       )


plt.title("Contoh Bar plot")
plt.xlabel("Ini sumbu X")
plt.ylabel("Ini sumbu Y")
plt.show()
```

```python
plt.figure(figsize= (10,7))
plt.bar(df1["x"], df1["z"],
       color = "green",
        linewidth= 3,
        edgecolor = "black"
       )

plt.bar(df1["x"], df1["y"],
       color = "orange",
        linewidth= 3,
        edgecolor = "black"
       )

plt.show()
```

```python
plt.figure(figsize= (10,7))
plt.bar(df1["x"]-.2, df1["z"],
       color = "green",
        width= .3,
        edgecolor = "black"
       )

plt.bar(df1["x"], df1["y"],
       color = "red",
        width= .3, ## 0.3
        edgecolor = "black"
       )

plt.show()
```

## 3. Scatter plot

```python
df2 = pd.DataFrame({
    "mesin": [1000, 1200, 1500, 3000, 5000],
    "harga": [6000, 7500, 8000, 9000, 12000]
})
```

```python
df2
```

```python
plt.scatter(df2["mesin"], df2["harga"],
           marker = "x",
           color = "red",
           s= 50) ## Size
plt.title("Korelasi antara Mesin dan Haarga")
plt.xlabel("Mesin")
plt.ylabel("Harga")
plt.grid(True)
plt.show()
```

## 4. Pie Chart

```python
presentase = [26, 44, 30]
nama = ["Mikiw", "Bagus", "Nisa"]

plt.pie(presentase, labels= nama,
       colors = ["orange", "green", "yellow"],
       startangle = 0, ## angle lingkaran
        textprops = {
            "size": 10,
            "color": "black"
        },
        autopct = "%1.2f%%",
        radius = 1, ## size lingkaran
        explode = (.02, .02, .02) ## garis putih/ pembatas
       )
plt.show()
```
