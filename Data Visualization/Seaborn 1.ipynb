## 1. Barplot

```python
plt.figure(figsize= (12,9))
sns.barplot(x= "sex", y= "total_bill", data= tips)

plt.show()
```

```python
plt.figure(figsize= (12,9))
sns.barplot(x= tips["day"], y= tips["total_bill"], estimator= np.std)
```
