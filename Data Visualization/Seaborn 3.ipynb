### pairplot

```python
sns.pairplot(df_taxi)
```

```python
sns.pairplot(df_taxi, hue = "payment", palette = "rainbow")
```

### PairGrid

```python
pg_taxi = sns.PairGrid(df_taxi);
pg_taxi
```

```python
p = sns.PairGrid(df_taxi)
p.map(plt.scatter);
```

```python
p = sns.PairGrid(df_taxi)
p.map_diag(sns.distplot)
p.map_upper(plt.scatter)
p.map_lower(sns.kdeplot)
```
